```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV
data_file="Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
new_coders_df=pd.read_csv(data_file,encoding='iso-8859-1',low_memory=False)
new_coders_df.head()


```

```python
# Inspect all columns
new_coders_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
ch_columns_coders_df=new_coders_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
ch_columns_coders_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
ch_columns_coders_df=ch_columns_coders_df.replace({0.0:"No", 1.0:"Yes"})
ch_columns_coders_df.head()
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
total_surveyed_df=len(ch_columns_coders_df)
total_surveyed_df
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
attended_bootcamp_df=ch_columns_coders_df.loc[ch_columns_coders_df["AttendedBootcamp"]=="Yes"]
attended_bootcamp_df
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
average_age = attended_bootcamp_df["Age"].mean()
average_age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
amount_attended_df=attended_bootcamp_df["AttendedBootcamp"].count()
amount_attended_df
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column
hold_degree_df=attended_bootcamp_df["SchoolDegree"].value_counts()
hold_degree_df
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column



total_degree=hold_degree_df.loc[["bachelor's degree","master's degree (non-professional)","professional degree (MBA, MD, JD, etc.)","associate's degree","Ph.D." ]].sum()
total_degree

```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

total_gender=attended_bootcamp_df["Gender"].value_counts().sum()

male=attended_bootcamp_df["Gender"].value_counts()["male"]

female=attended_bootcamp_df["Gender"].value_counts()["female"]

non_gender=(total_gender-male-female)




print(f"Total:{total_gender}\nMale: {male}\nFemale: {female}\nNon_Specific:{non_gender}\n")




```

```python
# Calculate percentage of respondents who attended a bootcamp
percent_attended=(amount_attended_df/total_surveyed_df)* 100
percent_attended

```

```python
# Calculate percentage of respondents belonging to each gender

total_gender=attended_bootcamp_df["Gender"].value_counts().sum()

male=attended_bootcamp_df["Gender"].value_counts()["male"]

female=attended_bootcamp_df["Gender"].value_counts()["female"]

non_gender=(total_gender-male-female)

male_percent=(male/total_gender) * 100
female_percent= (female/total_gender) * 100
non_gender_spec_percent=(non_gender/total_gender) * 100

print(f" % Male: {male_percent}\n % Female: {female_percent}\n % non_specifc: {non_gender_spec_percent}")



```

```python
# Calculate percentage of attendees with a college degree
degree_percentage=(total_degree/amount_attended_df) * 100
degree_percentage
```

```python
# Calculate average post-bootcamp salary
avg_salary= attended_bootcamp_df['BootcampPostSalary'].mean()
avg_salary

```

```python
# Create a new table consolidating above calculations
bootcamp_data = pd.DataFrame({"Total Surveyed": [total_surveyed_df],
                                   "Total Bootcamp attendees": [amount_attended_df],
                                   "% attended bootcamp": [percent_attended],
                                   "Avg. Age": [average_age],
                                   "% Male": [male_percent],
                                   "% Female": [female_percent],
                                   "% Non Gender Specific": [non_gender_spec_percent],
                                   "Has a degree": [degree_percentage],
                                   "Average Post Bootcamp Salary": [avg_salary]      })

bootcamp_data = bootcamp_data[["Total Surveyed",
                                         "Total Bootcamp attendees",
                                         "% attended bootcamp",
                                         "Avg. Age",
                                         "Has a degree",
                                         "% Male",
                                         "% Female",
                                         "% Non Gender Specific",
                                         "Average Post Bootcamp Salary"]]
bootcamp_data = bootcamp_data.round(2)

bootcamp_data




```

```python
# Improve formatting before outputting spreadsheet
bootcamp_data["% Male"] = bootcamp_data["% Male"].map("{0:,.2f}%".format)
bootcamp_data["% Female"] = bootcamp_data["% Female"].map("{0:,.2f}%".format)
bootcamp_data["% attended bootcamp"] = bootcamp_data["% attended bootcamp"].map("{0:,.2f}%".format)
bootcamp_data["% Non Gender Specific"] =bootcamp_data["% Non Gender Specific"].map("{0:,.2f}%".format)
bootcamp_data["Has a degree"] = bootcamp_data["Has a degree"].map("{0:,.2f}%".format)
bootcamp_data["Average Post Bootcamp Salary"] = bootcamp_data["Average Post Bootcamp Salary"].map("${0:,.0f}".format)

bootcamp_data


```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
bootcamp_data.to_excel("output/Bootcamppart1.xlsx", index=False)

```
